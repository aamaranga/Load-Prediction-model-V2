```python
import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('train_loan.csv')
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
#Loan_ID is useless.
df.drop(labels='Loan_ID',axis=1,inplace=True)
```

```python
for i in df.columns:
    if(df[i].dtype=='object'):
        fig, ax = plt.subplots()
        df[i].value_counts().plot(kind='bar',title=i)
```

```python
#create a copy of df to transform the qualitative data into numeric
df2 = df.copy()
df2.head()
```

```python
for i in df2.columns:
    if(df2[i].dtype!='object'):
        fig, ax = plt.subplots()
        df2[i].value_counts().plot(kind='hist',title=i)
```

```python
imputer = SimpleImputer(strategy='most_frequent')
le = LabelEncoder()
```

```python
for i in df2.columns:
    df2[i] = imputer.fit_transform(df2[i].values.reshape(-1,1))
```

```python
sc = StandardScaler()
df2.LoanAmount = sc.fit_transform(df2.LoanAmount.values.reshape(-1,1))
```

```python
for i in df2.columns:
    if(df2[i].dtype=='object'):
        df2[i] = le.fit_transform(df2[i].values.reshape(-1,1))
```

```python
sns.pairplot(df2)
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df2.corr(),annot=True)
```

```python
df2.LoanAmount.plot(kind='box')
```

```python
X = df2.drop(labels='Loan_Status',axis=1)
X = X.values
y =df2.Loan_Status.values
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score
```

```python
log_reg = LogisticRegression()
dt = DecisionTreeClassifier(max_depth=5)
nb = GaussianNB()
```

```python
clf = [log_reg,dt,nb]
score = []
for model in clf:
    score.append((model.__class__.__name__,cross_val_score(model,X,y)))
```

```python
for i in score:
        plt.plot(x=score[i][0],y=score[i][1])
```
